To fill missing values:

```python
calls["Purchase Amount"]=calls["Purchase Amount"].fillna(calls["Purchase Amount"].mean(),inplace=True)
```

```python
calls.isnull().sum()
```

```python
calls["Duration"]=calls["Duration"].fillna(calls["Duration"].mean(),inplace=True)
```

```python
calls["Satisfaction Rating"]=calls["Satisfaction Rating"].fillna(calls["Satisfaction Rating"].mean(),inplace=True)
```

To fill missing names:

```python
calls["Representative"]=calls["Representative"].fillna("Unknown",inplace=True)
```
