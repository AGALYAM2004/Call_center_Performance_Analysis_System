Extracting Month from Calls as a new column:

```python
calls['Month']=calls['Date of Call'].dt.month_name()
```

```python
calls.sample()
```

```python
Monthly_calls=calls.groupby("Month")["Call number"].count().sort_values(ascending=False)
print(Monthly_calls)
```

```python
Monthly_calls.plot(kind="line",marker="*")
plt.ylabel("No of calls")
plt.title("Monthly call Volume")
plt.show()
```

```python
Most_call_Rep=calls.groupby("Representative")["Call number"].count()
print(Most_call_Rep)
```

```python
Most_call_Rep.plot(kind="bar")
plt.ylabel("No of Calls")
plt.title("Rep Call Volume")
plt.show()
```

```python
Most_custo_rating=merged_table.groupby("Customer ID")["Rating"].sum()
print(Most_custo_rating)
```

```python
Most_custo_rating.plot(kind="bar")
plt.ylabel("Sum of Calls")
plt.title("Customers satisfaction Volume")
plt.show()
```
